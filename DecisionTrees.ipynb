```python
X = pd.get_dummies(df.drop('species',axis = 1),drop_first=True)
```

```python
y = df['species']
```

```python
X.head()
```

```python
y.head
```

```python
df['island'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
base_preds = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, mean_squared_error,classification_report,plot_confusion_matrix
```

```python
print(base_preds)
```

```python
print(y_test)
```

```python
classification_report(base_preds,y_test)
```

```python
confusion_matrix(base_preds,y_test)
```

```python
# Plotting the Tree
```

```python
from sklearn.tree import plot_tree
```

```python
fig = plt.figure(figsize = (12,8),dpi = 220)
plot_tree(model,feature_names=X.columns,filled=True)
```

```python
def report_model(model):
    preds = model.predict(X_test)
    print(classification_report(base_preds,y_test))
    print("\n")
    fig = plt.figure(figsize = (12,8),dpi = 220)
    plot_tree(model,feature_names=X.columns,filled=True)
```

```python
report_model(model)
```

```python
#Pruned Tree
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train,y_train)
```

```python
report_model(pruned_tree)
```

```python
 # NO. Of Nodes limited to 2 only.
```

```python
# Limiting Leafs :
max_leaf_model = DecisionTreeClassifier(max_leaf_nodes=3)
max_leaf_model.fit(X_train,y_train)
```

```python
report_model(max_leaf_model)
```

```python
# NO. of Leafs(Ends) limited to 3
```

```python
# Using Entropy in place of Gini
```

```python
ent_model = DecisionTreeClassifier(criterion="entropy")
```

```python
ent_model.fit(X_train,y_train)
```

```python
report_model(ent_model)
```

```python
min_leaves = DecisionTreeClassifier(min_samples_split=10) # Higher the value, more constrained tree.
```

```python
min_leaves.fit(X_train,y_train)
```

```python
report_model(min_leaves)
```

```python
min_leav_model = DecisionTreeClassifier(min_samples_leaf=10)
min_leav_model.fit(X_train,y_train)
```

```python
report_model(min_leav_model)
```

```python
two_model = DecisionTreeClassifier(max_leaf_nodes=2)
two_model.fit(X_train,y_train)
```

```python
report_model(two_model)
```
